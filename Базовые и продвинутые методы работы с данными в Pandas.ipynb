## 2.3. Работа с датами в DataFrame
### Признаки даты и времени

В наших данных об объектах недвижимости есть признак даты продажи - столбец Date. Из данного признака можно выделить множество полезной информации, например, год, месяц или день недели продажи имущества. На рынке недвижимости, как известно, присутствует сезонность: есть периоды, когда недвижимость покупается чаще, а есть интервалы времени, когда рынок претерпевает застой, поэтому было бы неплохо учитывать эту сезонность при анализе рынка.

#### Формат datetime

Единый способ обозначения даты и времени в Pandas - формат datetime, который записывается как YYYY-MM-DD HH: MM: SS, т.е составляющие времени указываются в следующем порядке: год, месяц, день, час, минута, секунда.
В наших данных дата записана в виде DD/MM/YYYY, например 3/12/2017. Посмотрим на это:

```python
display(melb_df['Date']) 
```

Для того, чтобы преобразовывать столбцы с датами, записанными в распространенных форматах, в формат datetime можно воспользоваться функцией pandas.to_datetime(). Преобразуем столбец Date в формат datetime передав его в эту функцию:

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
display(melb_df['Date'])
```

Стоит обратить внимание, что изменился тип данных для столбца Date, теперь его тип - datetime64. Рассмотрим, несколько возможностей данного типа данных.

### Выделение атрибутов datetime

Тип данных datetime позволяет с помощью специального аксессора dt выделять составляющие времени, такие как:

date - дата

year, month, day  - год, месяц, день

time - время

hour, minute, second - час, минута, секунда

dayofweek - номер дня недели, от 0 до 6, где 0 - понедельник, 6 - воскресенье

weekday_name - имя дня недели

dayofyear - порядковый день года

quarter - квартал (интервал в 3 месяца)

Замечание:

Аксессор - это атрибут столбца, который хранит переменные, которые были строковым представлением времени, а затем изменены с помощью pd.to_datetime(). Если вы попытаетесь обратиться к dt у столбца, в котором лежит что-то отличное от времени, вы получите ошибку.
Обратите внимание, что вы не сможете обратиться к аксессору, если ваш столбец не приведет к типу datetime.
Чуть больше можно увидеть здесь https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.html 

Например, обратившись по атрибуту dt.year к столбцу Date, мы можем достать год продажи и понять за какой интервал времени (в годах) у нас представлены данные, а также на какой год приходится наибольшее число продаж:

Из результатов видно, что данные представлены за интервал с 2016 по 2017 год и наибольшее количество объектов было продано в 2017 году.

Замечание:
В результате обращения к атрибуту datetime melb_data['Date'].dt.year мы получаем объект Series, в котором по значениям находятся годы продажи объектов недвижимости. ### 2.4. Создание и преобразование столбцов с помощью функций
Библиотека Pandas предоставляет большое количество возможностей для преобразований над данными, однако иногда необходимо совершать более сложные преобразования над столбцами. Например, из столбцов, содержащих в себе некоторый текст необходимо вытащить определенные слова, даты или числа. 
Для таких случаев Pandas не имеет специальных методов, но он предоставляет возможность расширить свою функциональность за счет использования пользовательских функций. 
Мы можем написать некоторую функцию, которая принимает на вход один элемент столбца, каким-то образом его обрабатывает и возвращает результат, после чего применить эту функцию к каждому элементу в столбце с помощью специального метода apply().
Рассмотрим пример: в наших данных есть столбец с адресами объектов недвижимости, проблема этого столбца в том, что в нем слишком большое количество уникальных значений: почти на каждый объект недвижимости в таблице приходится свой уникальный адрес. Убедимся в этом, вычислив количество уникальных значений в столбце с помощью метода nunique():

	
Такое большое количество возможных категорий может плохо сказаться на модели, которую мы бы хотели в дальнейшем построить и обучить на наших данных. Если не вдаваться в подробности, которые мы узнаем дальше в курсе, то из-за таких признаков зависимость между целевым признаком, который мы хотим предсказать, и признаками, на основе которых мы делаем предсказание, становится очень сложной, при этом точность моделирования при учитывании такого признака может даже вовсе не повыситься, а производительность однозначно резко упадет.
Выход из такой ситуации прост: давайте вытащим из признака адреса название улицы, тем самым, уменьшив количество уникальных адресов. Для этого сначала внимательнее посмотрим на структуру адреса в наших столбцах, выберем любую строку столбца Address:


```python
print(melb_df['Address'].nunique())
print(melb_df['Address'].loc[177])
```

Итак, наш адрес строится следующим образом: сначала указывается номер дома и корпус, а после указывается название улицы/шоссе/проспекта и т.д. Для того, чтобы выделить улицу, на которой находится объект, можно использовать следующую функцию:

```python
def get_street(address): #На вход  поступает строка с адресом
    address_list = address.split(' ')#1 split() разбивает строку на слова по пробелу и заносит их в список address_list
    address_list = address_list[1:] #2 обрезаем список, исключая из него слово, соответствующее номеру дома
    street = ' '.join(address_list) #3 оставшиеся слова объединяем по пробелу в одну строку с помощью метода join() 
    return street # и возвращаем полученную строку. 
```

А теперь применим эту функцию к столбцу c адресом: для этого передадим функцию get_street в аргумент метода столбца apply(), в результате получим объект Series, который положим в переменную streets:



```python
streets = melb_df['Address'].apply(get_street)
display(streets)
```

```python
#Мы смогли удалить из адреса номер дома и корпуса. Посмотрим, сколько уникальных значений у нас получилось:
print(streets.nunique())
```

### largest()

Итак, мы смогли уменьшить число адресов с 13371 до 5219 значений. Однако, этого все еще не достаточно. Тогда применим еще один очень распространенный метод уменьшения количества уникальных категорий: давайте выделим n улиц, которые встречаются чаще всего, а остальные улицы обозначим как ‘other’(другие).
Для этого сначала воспользуемся методом value_counts(), который вернет Series, значения которой -  частота каждой уникальной улицы. Теперь, когда мы знаем частоту для каждой улицы, мы можем выбрать 99 наиболее популярных (с наибольшей частотой) улиц, для можно воспользоваться методом nlargest(), который отбирает n наибольших значений в Series, в нашем случае n=99, наконец достанем наименования самых популярных улиц с помощью атрибута index. Результат занесем в переменную popular_streets:

```python
popular_streets = streets.value_counts().nlargest(99).index
print(popular_streets)
```

Теперь, когда у нас есть список наиболее популярных улиц, введем lambda функцию, которая будет проверять есть ли строка x в этом перечне, и, если это так, то она будет возвращать x, в противном случае она будет возвращать строку ‘other’. Наконец, применим такую функцию к Series streets, полученной ранее, а результат определим в столбец таблицы Streets:

```python
melb_df['Street'] = streets.apply(lambda x: x if x in popular_streets else 'other')
display(melb_df['Street'])
```

### Тип данных category

Для хранения и оптимизации работы с категориальными признаками в Pandas предусмотрен специальный тип данных category. Этот тип данных является гибридным: внешне он выглядит как строка, но внутренне представлен массивом целых чисел. Так как данные вместо изначальных строк хранятся в памяти как число, то объем памяти, занимаемой таблицей при использовании типа category резко уменьшается, что повышает эффективность хранения и работы с таблицей.
Более того, этот тип данных расширяет возможности работы с категориальными признаками: мы можем легко преобразовывать категории, строить графики по таким данным, а также резко повышается производительность операций, производимых с такими столбцами.

Самый простой способ преобразования столбцов к типу данных category - это использование уже знакомого нам метод astype(), в параметры которого достаточно передать строку ‘category’. Рассмотрим это на примере.
Для начала, выведем информацию о памяти, занимаемой текущей таблицей с помощью метода info():


```python
display(melb_df.info())
```

```python
melb_df.columns
```

```python
#Сделаем преобразование столбцов к типу данных category:
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] #список столбцов, которые мы не берем во внимание
max_unique_count = 150 #задаемся максимальным числом уникальных категорий
for col in melb_df.columns: #цикл по именам столбцов
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: #проверяем условие
        melb_df[col] = melb_df[col].astype('category') #преобразовываем тип столбца
display(melb_df.info())
```

Разберем код: сначала мы задаем список столбцов, которые мы не берем в рассмотрение, cols_to_exclude, а также условленный нами ранее порог уникальных значений столбца max_unique_count (первая и вторая строка кода). Далее в цикле перебираем имена столбцов, и, если число уникальных категорий меньше заданного порога и имя столбцов нет в списке cols_to_exclude, то с помощью метода astype() мы приводим столбец к типу данных category. 

Особенно хорошо такое преобразование работает на действительно больших данных, где число строк превышает сотни тысяч или миллионы. Иногда изменение типов может уменьшить объем памяти в десятки раз и существенно увеличить производительность.

### Сводные таблицы
сводная таблица на входе принимает данные из отдельных столбцов, группирует их и в результате получает новую таблицу, которая позволяет увидеть многомерное обобщение данных

#### Метод groupby как способ построения сводных таблиц
На самом деле мы с вами уже строили простейшие одномерные сводные таблицы с помощью метода groupby, мы рассматривали сводную таблицу в контексте группировки по одному признаку. Например, мы можем построить зависимость медианной цены и площади здания от числа комнат:


```python
melb_df.groupby(['Rooms'])[['Price', 'BuildingArea']].median()
```

Но можно пойти дальше и построить таблицу, в которой мы будем учитывать не только число комнат, но и тип здания (Type). В этот раз вместо среднего значения будем вычислять медиану. Для этого в параметрах метода groupby() мы указываем список из нескольких интересующих нас столбцов. Для того, чтобы финальный результат был представлен в виде таблицы, к результату агрегирования применяем метод unstack():


```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()
```

### Метод pivot_table для построения сводных таблиц
На самом деле метод groupby редко используется по 2ум параметрам, так как вместо него есть специальный и более простой метод pivot_table() для построения сводных таблиц. Разберем основные параметры данного метода:

values - имя столбца, по которому необходимо построить сводные данные, применяя агрегирующую функцию

index - имя столбца, значения которого станут строками сводной таблицы

columns - имя столбца, значения которого станут столбцами сводной таблицы

aggfunc - имя или список имен агрегирующих функций. По умолчанию подсчет среднего - ‘mean’.

fill_value - значение, которым заполнить пропуски. По умолчанию пропуски не заполняются.

```python
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type'
).round()
```

Несложно понять, что метод pivot_table() имеет преимущество перед группировкой по нескольким критериям. Оно заключается в наличии специальных аргументов для строк и столбцов сводной таблицы, благодаря чему уменьшается вероятность запутаться при построении более сложных (многомерных) сводных таблиц, о которых мы поговорим далее.
А теперь давайте произведем анализ продаж в каждом из регионов в зависимости от того, будний ли был день или выходной. Для этого построим сводную таблицу, в которой строками будут являться названия регионов (Regionname), а по столбцам будет стоять наш “признак-мигалка” выходного дня (Weekend), который равен 1, если день был выходным и 0 в противном случае. В качестве значений сводной таблицы возьмем количество продаж.

```python
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
)
```

Найдем, как зависит средняя и медианная площадь  участка (Landsize) от типа объекта (Type) и его региона (Regionname). Для того, чтобы посмотреть несколько статистических параметров, нужно передать в аргумент aggfunc список из агрегирующих функций. Построим такую сводную таблицу, где пропущенные значения заменим на 0:


```python
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0)
```

 сводные таблицы позволяют рассматривать зависимость и от большего числа признаков. Такие сводные таблицы уже называются многомерными. 
Для того, чтобы исследовать зависимость от большего числа признаков, можно передать список признаков в параметр index или параметр columns. Посмотрим это на примере. Построим таблицу, в которой по индексам отложим признаки метода продажи (Method) и типа объекта (Type), по столбцам - наименование региона (Regionname), а на пересечении строк и столбцов будет стоять медианная цена объекта (Price):


```python
melb_df.pivot_table(
    values='Price',
    index=['Method','Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

##### Доступ к данным в сводной таблице
Возникает вопрос, а как получить доступ к данным или произвести фильтрацию в сложной сводной таблице, где есть дополнительные индексы? Давайте посмотрим на то, что представляют из себя столбцы сложной сводной таблицы.
Возьмем следующую сводную таблицу, которую мы создавали ранее:

И выведем ее столбцы с помощью атрибута columns:


```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

```python
print(pivot.columns)
```

В результате мы получаем объект MultiIndex (мульти индекс). Этот объект хранит в себе 6 комбинаций пар столбцов (2 статистических параметра и  3 типа здания). Т.е есть 6 возможных вариантов обращения к столбцам таблицы.
На самом деле мульти индексы раскрываются подобно вложенным словарям: по очереди, как матрешка. Для того, чтобы получить доступ к определенному столбцу, вы должны сначала обратиться к столбцу, который находится уровнем выше. Так, из таблицы pivot мы можем получить средние значения площадей участков для типа здания unit просто последовательно обратившись по имени столбцов:


```python
display(pivot['mean']['unit'])
```

##### Фильтрация в сводных таблицах
Аналогично производится и фильтрация в данных. Например, если нам нужны регионы, в которых средняя площадь здания для домов типа house меньше их медианной площади, то мы можем это сделать следующим образом:


```python
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
display(filtered_pivot)
```
